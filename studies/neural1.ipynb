```python
from sklearn import datasets
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
rna_input = pd.read_csv('rna_input2.csv')
rna_input = rna_input.dropna()
```

```python
old_rna = pd.read_csv('rna_input.csv')
```

```python
old_rna
```

### rna_input = rna_input.drop(['Unnamed: 0', 'index'], axis=1)


```python
rna_input
```

```python
x = rna_input[['Barreto_ws10','Barreto_wd10','prec_prev_Barreto', 'temp_prev_Barreto']]
y = rna_input['precipitacao_observada']
rna = MLPRegressor(hidden_layer_sizes=(10,5), max_iter=20000
, learning_rate='constant', learning_rate_init=0.001, solver='sgd', activation='logistic'
, tol=0.00000001, verbose=2)
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3, random_state=1)
rna.fit(x_treino, y_treino)
```

```python
pred = rna.predict(x_teste)
pred
```

```python
y_teste
```

```python
(y_teste - pred).mean()
```
