## Vai funcionar, mas antes de concatenar a segunda variável, tem que nomear as novas colunas do primeiro merge

```python
dates = pd.to_datetime(conc2['Datetime'])
```

```python
first = dates.iloc[0]
last = dates.iloc[-1]
```

```python

import datetime
missing_dates = []
presente_dates = []
while first <= last:
    if first not in list(dates.values):
        missing_dates.append(first)
    else:
        presente_dates.append(first)

    first += datetime.timedelta(hours=1)
```
